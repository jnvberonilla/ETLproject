```python
import os
import csv
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm import Session

#import methods for abstracting classes into tables
from sqlalchemy.ext.declarative import declarative_base
#import methods for declaring column types
from sqlalchemy import Column, Integer, String, Float, Numeric

engine = create_engine('sqlite:///drunks.sqlite', echo="debug")
conn = engine.connect()
Base = declarative_base()
# Base = automap_base()
```

```python
consumption = "csvData.csv"
consumption_df = pd.read_csv(consumption)
consumption_df.rename(columns={'State': 'state', 'alcoholConsumptionGallons': 'volume'}, inplace=True)
consumption_df
```

```python
impaired = "Impaired_Driving_Death_Rate__by_Age_and_Gender__2012___2014__All_States.csv"
impaired_df = pd.read_csv(impaired)
impaired_df
```

```python
#impaired_clean = impaired_df.drop(['Location', 'All Ages, 2014', 'Ages 0-20, 2012', 'Ages 0-20, 2014 '])
impaired_clean = impaired_df.drop(columns=['Location',
                                           'All Ages, 2014', 
                                           'Ages 0-20, 2012', 
                                           'Ages 0-20, 2014', 
                                           'Ages 21-34, 2012',
                                           'Ages 21-34, 2014', 
                                           'Ages 35+, 2012', 
                                           'Ages 35+, 2014', 
                                           'Male, 2012', 
                                           'Male, 2014', 
                                           'Female, 2012', 
                                           'Female, 2014'])
impaired_clean.rename(columns={'State ': 'state', 'All Ages, 2012': 'rate'}, inplace=True)
impaired_clean
```

```python
# Creates Classes which will serve as the anchor points for our Tables
class Impaired(Base):
    __tablename__ = 'impaired'
    id = Column(Integer, primary_key=True)
    state = Column(String(255))
    rate = Column(Numeric)
class Consumption(Base):
    __tablename__ = 'consumption'
    id = Column(Integer, primary_key=True)
    state = Column(String(255))
    volume = Column(Numeric)
Base.metadata.create_all(engine)
```

```python

# Add Records to the Appropriate DB
# ----------------------------------
# Create a Session Object to Connect to DB
session = Session(engine)
# loop through impaired_clean and consumption_df data frames to fill database
for index, row in impaired_clean.iterrows():
    impaired = Impaired(state =row[0], rate=row[1])
    session.add(impaired)
    
for index, row in consumption_df.iterrows():
    consumption = Consumption(state=row[0], volume=row[1])
    session.add(consumption)
    
session.commit()
```

```python
# Query the Tables
# ----------------------------------
# Perform a simple query of the database
impaired_list = session.query(Impaired)
for i in impaired_list:
    print(i.rate)
consumption_list = session.query(Consumption)
for c in consumption_list:
    print(c.volume)
```
